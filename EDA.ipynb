```python
categorical_columns=['Gender','Marital_status','Number_of_dependents','Housing','Housing','Housing','Housing']
```

```python
fig,axes = plt.subplots(4,2,figsize=(12,15))
for index,cat_col in enumerate(categorical_columns):
    row,col = index//2,index%2
    sns.countplot(x=cat_col,data=data1,hue='Gender',ax=axes[row,col])
```

```python
categorical_column2=['Purpose','Property','Has_coapplicant','Has_guarantor','Number_of_existing_loans_at_this_bank','Loan_history']
```

```python
fig,axes = plt.subplots(3,2,figsize=(20,15))
for index,cat_col in enumerate(categorical_column2):
    row,col = index//2,index%2
    sns.countplot(y=cat_col,data=data2,hue='high_risk_applicant',ax=axes[row,col])
```
